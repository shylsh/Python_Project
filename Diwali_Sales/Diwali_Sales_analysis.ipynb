## Sales Analysis

```python
# import Libraries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')
df.head()
```

```python
df.shape
```

```python
df.tail()
```

```python
df.info()
```

```python
df.drop(columns=['Status', 'Unnamed'], axis=1, inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
rows_with_nulls = df[df.isnull().any(axis=1)]
```

```python
print(rows_with_nulls)
```

```python
df['Amount'] = df.groupby('Product_Category')['Amount'].transform(lambda x: x.fillna(x.mean()))
```

```python
df['Amount'] =df['Amount'].astype(int)
```

```python
df.info()
```

```python
df.columns
```

```python
df.rename(columns={'Marital_Status': 'Married'})
```

```python
def categorize_age(age):
    if age <= 17:
        return '0-17'
    elif age <= 25:
        return '18-25'
    elif age <= 35:
        return '26-35'
    elif age <= 45:
        return '36-45'
    elif age <= 55:
        return '46-55'
    else:
        return '56+'
```

```python
df['Age_Group'] = df['Age'].apply(categorize_age)
```

# EDA

```python
ax = sns.countplot(x = 'Gender', data =df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)
plt.show()
```

### Age

```python
ax = sns.countplot(data = df , x = 'Age_Group', hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen1 = df.groupby(['Age_Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x = 'Age_Group', y = 'Amount', data = sales_gen)
plt.show()
```

```python
sales_state = df.groupby(['State'],as_index = False)['Orders'].sum().sort_values(by = "Orders",ascending=False).head(5)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y = 'Orders')
plt.show()
```

### Marital Status

```python
ax = sns.countplot(data = df, x = 'Marital_Status')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Marital_Status','Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.barplot(x = 'Marital_Status', y = 'Amount', data = sales_gen, hue = 'Gender')
plt.show()
```
